#Getting the Baby Names data from the SSA

This data now extends up to 2012.


```python
import urllib2
from cStringIO import StringIO

try:
   zf = open("names.zip","r")     #  try to read in a cached names.zip file, if it is there
   zfileString = StringIO(zf.read())
   zfiledata = zfileString.getvalue()
except:
   zipped_data = urllib2.urlopen("http://www.ssa.gov/oact/babynames/names.zip")
   zfileString = StringIO(zipped_data.read())   #  get the (large) .zip file from SSA
   zfiledata = zfileString.getvalue()
   try:
      zf = open("names.zip","w")  # cache the .zip file here for next time
      zf.write(zfiledata)
      zf.close()
   except:
      pass

print "read in SSA names.zip file; length = %d" % len(zfiledata)
```

```python
import zipfile
zfile = zipfile.ZipFile(zfileString)
print zfile.namelist()  # list of all files in the zip file
```

```python
pieces = []
columns = ['name', 'sex', 'births']

#  process the .txt (=.csv) file for each year 1880 - 2012.

for filename in sorted(zfile.namelist()):
   if filename.endswith(".txt"):  #  load each .txt file within the .zip file
      try:
         year = int(filename[3:7])
      except:
         year = 1880
      frame = pd.read_csv(zfile.open(filename), names=columns)
      frame['year'] = year        #  add a 'year' column to this dataframe
      pieces.append(frame)        #  append this year's dataframe to a list

# Concatenate everything into a single DataFrame
SSA_Baby_Names = pd.concat(pieces, ignore_index=True)

SSA_Baby_Names.head()             #  print the first few rows of the dataframe
```

```python
SSA_Baby_Names.info()      #  print the dataframe schema

SSA_Baby_Names.describe()  #  summary statistics for numeric columns
```

```python
total_births = names.pivot_table('births', rows='year', cols='sex', aggfunc=sum)
total_births.tail()
```

```python
def add_prop(group):
    births = group.births.astype(float)
    group['prop'] = births / births.sum()
    return group

names = names.groupby(['year', 'sex']).apply(add_prop)

names
```

```python
np.allclose(names.groupby(['year', 'sex']).prop.sum(), 1)
```

```python
def get_top1000(group):
    return group.sort_index(by='births', ascending=False)[:1000]
```

```python
grouped = names.groupby(['year', 'sex'])
top1000 = grouped.apply(get_top1000)
```

### Boy names that became girl names (and vice versa)

```python
all_names = top1000.name.unique()
mask = np.array(['lesl' in x.lower() for x in all_names])
lesley_like = all_names[mask]
lesley_like
```

```python
filtered = top1000[top1000.name.isin(lesley_like)]
filtered.groupby('name').births.sum()
```

```python
table = filtered.pivot_table('births', rows='year', cols='sex', aggfunc='sum').fillna(0)
table = table.div(table.sum(1), axis=0)
table.tail()
```

```python
table.plot(style={'M': 'k-', 'F': 'k--'}, title='Proportion of male/female Lesley-like names over time')
```
