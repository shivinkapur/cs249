### Modeling the meme life-cycle pattern

1. 
We try to model using the saw-tooth + exponential distribution given in the Meme-Tracking paper i.e.

For a name,
X= vector of number of births with that name
t= year

$  X= A t^B e^{-Ct}  $

We perform optimizing this equation for Amy using optim() and minimising the squared error. To go about doing this, we first normalize X vector between (0,1) and shift the peak of the life-cycle at t=0. 
On running optim() several times with different combinations of initial parameters as well as trials on different names, we do not get a good fit. 



```python
%%R  -w 900 -h 900

amy= subset(dataset, dataset$name %in% c("Amy"))
print(head(amy))

amyDataset= data.frame(amy$shiftYear, amy$normBirths)
colnames(amyDataset)=c("shiftYear", "normBirths")

#using saw-tooth to fit model in data and optimizing for minimum error
min.RSE= function(data, par){
    with(data, sum((par[1]*((shiftYear)^par[2])*exp(-(shiftYear))- normBirths)^2))
    }


result_amy= optim(par=c(0.05, 20), min.RSE, data=amyDataset)
print(result_amy)

funx=function(x){result_amy$par[1]*((x)^(result_amy$par[2]))*exp(-(x))}
print(plot(amy$shiftYear, amy$normBirths))
curve(funx,-95 , 37, add=TRUE, col="royalblue", lwd=2)

#does not give a good fit
cat()
```

Since optim() does not give a good fit due to our many attempts, we try finding the parameter values by performing least squares regression. 

We rewrite $  X= A t^B e^{-Ct}  $  as  

$ log{X}= log{A} + B log{t} - Ct $ 

$ log{X}= A' + B log{t} - Ct $

We solve the above equation using regression to compute coefficients A', B and C. As depicted in the last plot of the cell below, the model doesn't fit well as the coefficient C completely diminishes the value of t. Thus it only shows the saw-tooth increase but doesn't model the exponential decrease. 

```python
%%R  -w 900 -h 900

#Trying to solve for parameters using least squares method 
amy.logX= log(amyDataset$normBirths)
Y=amy.logX
amy.logT=log(amy$shiftYear+132)
amy.minusT=-(amy$shiftYear+132)

lseModel= lm(Y ~ amy.logT + amy.minusT )
print(summary(lseModel))

plot(amyDataset$normBirths)

v=-13.65 + 1.15*amy.logT - 0.02*amy.minusT

pred= predict(lseModel)

plot(amy.logX)
points(v, col="red")
```

By using a smoothing spline, what one really wants to do is to find a function which fits the observed data superbly well while keeping the RSS as small as possible. Also, we don't want to make RSS zero since that will effectively overfit the data. A smoothing spline is simply a natural cubic spline with knots at every unique value of a datapoint.

So by using smoothing splines, we want to minimize the loss but keep a penalty as a measure of its roughness. A tuning parameter lamda; associated with the penalty term; can help us do so. Lamda control the spline's effective degree of freedom. The value of lamda which is chosen for a smoothing spline is one that makes the cross-validated RSS as small as possible. 



#### Spline can not be used for modeling the meme life-cycle pattern since it is difficult to find a value of lambda which can be used to make a model for each name.

#### 3. Using different distributions (weibull, gamma, lognormal) to model names 

```python
%%R

library(MASS)

evaluateParams= function(n){

print(n)
x=subset(dataset, dataset$name %in% c(n))

freq=x$births
years=x$year
yearsFreq=rep(years, freq)

x.param.weibull = fitdistr( yearsFreq, "weibull" )
x.param.gamma = fitdistr(yearsFreq, "gamma" )
x.param.lognormal = fitdistr(yearsFreq, "lognormal")

    
print(x.param.weibull)
print(x.param.weibull$loglik)
print(x.param.gamma)
print(x.param.gamma$loglik)
print(x.param.lognormal)
print(x.param.lognormal$loglik)

}

namesToUse= c("Amy", "Gladys",  "Gloria", "Gary", "Harold",  "Joshua",  "Linda",  "Michelle", "Terry",  "Trinity", "Bruce", "Aaliyah")
namesToUse=as.character(namesToUse)
for(i in 1:length(namesToUse)){
    evaluateParams(namesToUse[i])
}
```

#### We find that lognormal performs best for most and thus we proceed to model the names using lognormal function i.e.  $ f(x) = \frac{e^{-((\ln x)^{2}/2\sigma^{2})}}  {x\sigma\sqrt{2\pi}} \hspace{.2in} x > 0; \sigma > 0 $ x is the number of births over the name's life-cycle (here we fit the distribution on yearFreq vector where each year is repeated as a factor of the number of occurences)

```python
%%R

#Evaluating log likelihood for all 1567 unique names having only one max peak 


finalEval= function(n){

#print(n)
x=subset(dataset, dataset$name %in% c(n))
freq=x$births
years=x$year
yearsFreq=rep(years, freq)
#print(dim(x))
#print(head(x))
    
x.param.lognormal = fitdistr(yearsFreq, "lognormal")

meanlog=x.param.lognormal$estimate[1]
sdlog= x.param.lognormal$estimate[2]
loglik= x.param.lognormal$loglik

l=c(n, meanlog, sdlog, loglik)
return(l)
}

namesToUse= unique(dataset$name)
namesToUse=as.character(namesToUse)
print(head(namesToUse))
print(mode(namesToUse))

N <- 1e4
fullListParams <- data.frame(name=rep("", N), meanlog=rep(NA, N), sdlog=rep(NA, N), loglik=rep(NA, N),
                 stringsAsFactors=FALSE) 

for(i in 1:length(namesToUse)){
    fullListParams[i,]= finalEval(namesToUse[i])
}

cat()
```

#### We evaluate the fit of the model based on the log likelihood measure 

```python
%%R

print(head(fullListParams))
fullListParams=fullListParams[1:1567,]
print(dim(fullListParams))

fullListParams$meanlog=as.numeric(fullListParams$meanlog)
fullListParams$sdlog=as.numeric(fullListParams$sdlog)
fullListParams$loglik=as.numeric(fullListParams$loglik)
print(summary(fullListParams))

hist(fullListParams$loglik)

#Computing the 95% confidence interval of the log likelihood measure 
library(Rmisc)
loglik.confint= CI(fullListParams$loglik, ci=0.95)
print(loglik.confint)


#Plotting the models for names: Aaliyah and Aaron
#Loglikelihood measure for Aaliyah is worse than that of Aaron's, as visible from the histograms
aaliyah=subset(dataset, dataset$name %in% c("Aaliyah"))
plot(aaliyah$year, aaliyah$normBirths, main="Aaliyah")
freq=aaliyah$births
years=aaliyah$year
yearsFreq=rep(years, freq)
hist( yearsFreq, probability=TRUE, col="orange", border=NA, main="Aaliyah")
parameters10 = fitdistr(yearsFreq, "lognormal")
curve(dlnorm(x, meanlog=parameters10$estimate[1], sdlog=parameters10$estimate[2]), col= "black", add=TRUE )
legend("right", c(paste("Lognormal Distribution")), col=c("black"), lwd=5 )

aaron=subset(dataset, dataset$name %in% c("Aaron"))
plot(aaron$year, aaron$normBirths, main="Aaron")
freq=aaron$births
years=aaron$year
yearsFreq=rep(years, freq)
hist( yearsFreq, probability=TRUE, col="orange", border=NA, main="Aaron"  )
parameters10 = fitdistr(yearsFreq, "lognormal")
curve(dlnorm(x, meanlog=parameters10$estimate[1], sdlog=parameters10$estimate[2]), col= "black", add=TRUE )
legend("right", c(paste("Lognormal Distribution")), col=c("black"), lwd=5 )

#plotting model for abbie, we can see that the lognormal model doesn't fit well for abbie
abbie=subset(dataset, dataset$name %in% c("Abbie"))
plot(abbie$year, abbie$normBirths, main="Abbie")
freq=abbie$births
years=abbie$year
yearsFreq=rep(years, freq)
hist( yearsFreq, probability=TRUE, col="orange", border=NA, main="Abbie"  )
parameters10 = fitdistr(yearsFreq, "lognormal")
curve(dlnorm(x, meanlog=parameters10$estimate[1], sdlog=parameters10$estimate[2]), col= "black", add=TRUE )
legend("right", c(paste("Lognormal Distribution")), col=c("black"), lwd=5 )



minLoglikName=subset(fullListParams, loglik < (-8000000))$name
print(minLoglikName)

#these look like best examples for our model
christopher=subset(dataset, dataset$name %in% c("Christopher"))
plot(christopher$year, christopher$normBirths, main="Christopher")
freq=christopher$births
years=christopher$year
yearsFreq=rep(years, freq)
hist( yearsFreq, probability=TRUE, col="orange", border=NA, main="Christopher"  )
parameters10 = fitdistr(yearsFreq, "lognormal")
curve(dlnorm(x, meanlog=parameters10$estimate[1], sdlog=parameters10$estimate[2]), col= "black", add=TRUE )
legend("right", c(paste("Lognormal Distribution")), col=c("black"), lwd=5 )
```
