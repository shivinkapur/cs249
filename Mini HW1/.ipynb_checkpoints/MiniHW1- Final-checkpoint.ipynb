## Problem 1.  What <i>kind</i> of distribution does the number of Beneficiaries appear to follow?

```python
%%R -w 600 -h 400

library(MASS)
NumberOfBeneficiaries = NumericData[,2]
# intuitively, it would make sense if this number followed some kind of arrival distribution

# The number of beneficiaries is very large in some cases (see the summary above),
# so let's try a plot of log10(number of beneficiaries)
# to see how many there are

hist( log10(NumberOfBeneficiaries), probability=TRUE, col="deepskyblue", breaks=50 )
parameters1 = fitdistr( log10(NumberOfBeneficiaries), "weibull" )
curve( dweibull(x, shape=parameters1$estimate[1], scale=parameters1$estimate[2]), 1, 4, col="red", add=TRUE, ylim=c(0,6000) )  # fits relatively very well

legend("right", c(paste("Weibull Distribution")), col=c("red"), lwd=5 )
```

```python
%%R
# The histogram above shows that most numbers of beneficiaries are below 10^(3.7) = 5,000.

# The R expression  (NumberOfBeneficiaries < 5000)  yields a boolean vector.
# It can be used as a subscript to select all entries of a vector satisfying this boolean:

NumberOfBeneficiariesBelow5K = NumberOfBeneficiaries[ (NumberOfBeneficiaries < 5000) ]
print(head(NumberOfBeneficiariesBelow5K))
hist( NumberOfBeneficiariesBelow5K, probability=TRUE, col="yellow", border=NA, breaks=500  )

parameters10 = fitdistr(NumberOfBeneficiariesBelow5K, "lognormal")
curve(dlnorm(x, meanlog=parameters10$estimate[1], sdlog=parameters10$estimate[2]), col= "black", add=TRUE )

legend("right", c(paste("Lognormal Distribution")), col=c("black"), lwd=5 )
```

## inspect the distributions of Charges, Allowed Amounts, and Payment Amounts

```python
%%R -h 800 -w 800

# First, health care providers _charge_ Medicare.
# Then Medicare determines what charges are _allowed_.
# Eventually Medicare actually _pays_ the providers.

SubmittedChargeAmounts = NumericData[,4]
MedicareAllowedAmounts = NumericData[,5]
MedicarePaymentAmounts = NumericData[,6]
                               
opar = par(mfrow=c(2,2))
 
hist( log10(SubmittedChargeAmounts), col="red", probability=TRUE, border=NA, breaks=500 )
parameters1_1 = fitdistr( log10(SubmittedChargeAmounts), "weibull" )
curve( dweibull(x, shape=parameters1_1$estimate[1], scale=parameters1_1$estimate[2]), col="navyblue", add=TRUE )  # fits relatively very well
legend("right", c(paste("Weibull Distribution")), col=c("navyblue"), lwd=5 )


hist( log10(MedicareAllowedAmounts), col="orange", probability=TRUE, border=NA, breaks=500 )
parameters2_1 = fitdistr( log10(MedicareAllowedAmounts), "weibull" )
curve( dweibull(x, shape=parameters2_1$estimate[1], scale=parameters2_1$estimate[2]), col="navyblue", add=TRUE )  # fits relatively very well
legend("right", c(paste("Weibull Distribution")), col=c("navyblue"), lwd=5 )


MedicarePaymentAmountsAboveOne = MedicarePaymentAmounts[(MedicarePaymentAmounts > 1)]
hist( log10(MedicarePaymentAmountsAboveOne), col="green3", probability=TRUE, border=NA, breaks=500 )

parameters3_1 = fitdistr( log10(MedicarePaymentAmountsAboveOne), "weibull" )
curve( dweibull(x, shape=parameters3_1$estimate[1], scale=parameters3_1$estimate[2]), col="navyblue", add=TRUE )  # fits relatively very well
legend("right", c(paste("Weibull Distribution")), col=c("navyblue"), lwd=5 )

# Also, look at the difference between the Submitted Charges and the Actual Payments:

DifferencePaymentAmounts = SubmittedChargeAmounts - MedicarePaymentAmounts
DifferencePaymentAmountsAboveOne = DifferencePaymentAmounts[(DifferencePaymentAmounts > 1)]

hist( log10( DifferencePaymentAmountsAboveOne ), col="royalblue", probability=TRUE, border=NA, breaks=500 )
parameters4_1 = fitdistr( log10(DifferencePaymentAmountsAboveOne), "weibull" )
curve( dweibull(x, shape=parameters4_1$estimate[1], scale=parameters4_1$estimate[2]), col="pink", add=TRUE )  # fits relatively very well
legend("right", c(paste("Weibull Distribution")), col=c("pink"), lwd=5 )


par(opar)
```

## Problem 3.  Fit a distribution to: log10( average charge amounts )

```python
%%R

AverageChargeAmounts =  log10( SubmittedChargeAmounts / NumberOfBeneficiaries )


average_amounts_histogram = hist( AverageChargeAmounts, 
                                 probability=TRUE, border=NA, col="cyan", breaks=500 )

mtext("this looks like a mixture, but you can ignore the amounts below 1.7", side=3, col="blue")

AverageChargeAmounts = AverageChargeAmounts[ AverageChargeAmounts >= 1.7 ]

# what is a good fit for the distribution of AverageChargeAmounts? (ignoring the bump at the low end)

# try using fitdistr() to approximate a fit

parameters4 = fitdistr( AverageChargeAmounts, "logistic" )
print(parameters4)
curve( dlogis(x, location=parameters4$estimate[1], scale=parameters4$estimate[2]), col="black", add=TRUE ) #fits rather well

mtext( "The logistic distribution fits the best!", line=-1, side=3, col="black" )
legend("right", c(paste("Logistic Distribution")), col=c("black"), lwd=5 )
```

## Problem 5.  Does Total Charge Amount by State follow a Power Law?

```python
%%R -w 800 -h 800

total_charge_by_state = aggregate( total_submitted_chrg_amt  ~  nppes_provider_state,  data=MedicarePaymentData,  sum  )

hist( log( total_charge_by_state[,2] ), col="seagreen", breaks=20 )

ranked_by_state = total_charge_by_state[ order( total_charge_by_state[,2], decreasing=TRUE ), ]

# keep only the top 50 states (ignore international data)
ranked_by_state = ranked_by_state[1:50,]

ranks = 1:nrow(ranked_by_state)


plot(   log(ranks), log(ranked_by_state$total_submitted_chrg_amt), probability=TRUE, type="l", col="blue" )  # line segments first
#curve( dexp(x, rate=5.0), 0,4, ylim=c(18,24), col="yellow", add=TRUE)

points( log(ranks), log(ranked_by_state$total_submitted_chrg_amt),col="darkblue", pch=20 )  # add dots

text( log(ranks), log(ranked_by_state$total_submitted_chrg_amt),
     ranked_by_state$nppes_provider_state, pos=2, cex=0.75, col="red" )

# ... is this relationship a power law?  Is the plot essentially log-log-linear?
# (Does it look like "the rich get richer" ?)

print( "Power law distributions are in general of the form f(k) = a/k^c. Taking logarithm on both sides we get logf(k) = loga−c logk. This says that if we have a power-law relationship, and we plot logf(k) as a function of logk, then we should see a straight line: −c will be the slope, and loga will be the y-intercept. But since the plot below isn't linear hence total charge amount by state doesn't follow a Power Law" )

```

## Problem 5.  What is the distribution of maximum charges made by the different kinds of Providers?

```python
%%R -w 1100

max_charge_by_provider_type = aggregate( total_submitted_chrg_amt  ~  provider_type,  data=MedicarePaymentData,  max  )

hist( log10( max_charge_by_provider_type$total_submitted_chrg_amt ), col="purple", probability=TRUE)

parameters = fitdistr( log10(max_charge_by_provider_type$total_submitted_chrg_amt), "logistic" )
print(parameters)
curve( dlogis(x, location=parameters$estimate[1], scale=parameters$estimate[2]), col="black", add=TRUE ) #fits rather well

legend("right", c(paste("Logistic Distribution")), col=c("black"), lwd=5 )


rank_ordering_by_max_charge = order( log10(max_charge_by_provider_type[,2]), decreasing=TRUE )

ranked_by_type = max_charge_by_provider_type[ rank_ordering_by_max_charge , ]

print(ranked_by_type)

ranks = 1:nrow(ranked_by_type)

plot(   log(ranks), log(ranked_by_type$total_submitted_chrg_amt), type="l", col="blue" )
points( log(ranks), log(ranked_by_type$total_submitted_chrg_amt), col="darkblue", pch=20 )

text( log(ranks), log(ranked_by_type$total_submitted_chrg_amt),
     ranked_by_type$provider_type, pos=2, cex=0.5, col="red" )
```
