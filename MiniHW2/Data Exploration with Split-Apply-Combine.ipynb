##Load in the Baby Names data

```python
%%R

bnames <- read.csv("baby-names.csv", stringsAsFactors = FALSE)

print(dim(bnames))

print(summary(bnames))

cat("\n\n first few and last few rows in the baby names table:\nthere are 129 years * 1000 top names/year * 2 genders = 258K rows\n\n")

print(head(bnames))

print(tail(bnames))
```

##Doing things the wrong way -- with for-loops

The natural instinct is to do things with for-loops.  Avoid this instinct -- you are not only slowing yourself down, you are likely to commit more errors, and your code will be harder change or improve.

```python
%%R

#  problem:  construct a table that shows, for each gender, and for each year, the total of all percent values.

genders = c("girl", "boy")

years = 1880:2008
nyears = length(years)

results = data.frame(
    gender = c( rep("girl",nyears), rep("boy",nyears) ),
    year = c(years,years) )

totals = rep(0,2*nyears)

i = 1
for (gender in genders) {
   for (year in years) {
      percents = bnames$percent[ ((bnames$year == year) & (bnames$sex == gender)) ]
      totals[i] = sum(percents)
      i = i+1
  }
}

results$total_percent = totals

print(head(results))

print(tail(results))
```

```python
%%R

# now plot the results

girl_percents = results[ results$gender == "girl", ]$total_percent

boy_percents  = results[ results$gender == "boy", ]$total_percent

plot( years, girl_percents, col="deeppink", type="l", lwd=1.5,
     xlab="year", ylab="percent", main="doing things the hard way" )
points( years, boy_percents, col="skyblue", type="l", lwd=1.5 )

mtext( "rats, I should have plotted boys first, their plot is clipped now",
      side=3, col="blue")
```

```python
%%R

# a better way of obtaining subsets in R:  the  "subset"  function

girl_results = subset( results, gender == "girl" )

boy_results  = subset( results, gender == "boy" )

plot( years, girl_results$total_percent, col="deeppink", type="l", lwd=1.5,     ylim = c(0.5, 1.0),
     xlab="year", ylab="percent", main="doing things a slightly better way" )

points( years, boy_results$total_percent, col="skyblue", type="l", lwd=1.5 )

```

```python
%%R

#  But wait, there's more!  R also has a 'split' function that divides data into groups:

gender_results = split( results, as.factor(results$gender))

print(str(gender_results))

plot( years, gender_results$girl$total_percent, col="deeppink", type="l", lwd=1.5, ylim = c(0.5, 1.0),
     xlab="year", ylab="percent", main="better still" )

points( years, gender_results$boy$total_percent, col="skyblue", type="l", lwd=1.5 )
```

##Doing things in an almost respectable way

R has lots of aggregate functions, like aggregate functions in SQL, notably:

---  aggregate( data.frame, groupby_defining_vector, function_to_apply_on_groups )
  
---  by( data, groupby_defining_vector, function_to_apply_on_groups )
  
  

```python
%%R

percentages = aggregate( bnames$percent, list( bnames$year, bnames$sex ), sum )

## by_percentages = by( bnames$percent, list( bnames$year, bnames$sex ), sum )
## ## this is similar but the result is a 'by' structure, incorporating grouping 


 ## notice that the reult of 'aggregate' is a data.frame:
print( head(percentages) ) 
```

###R also has many versions of "apply", which apply a function to a structure:

--- apply( Array, (1 -> row-wise, 2 -> column-wise), Function_to_apply_to_each_row_or_each_column )
     #  maps an Array to an Array or List of values

--- tapply( Vector, groupby_defining_vector, Function_to_apply_to_each_group )
     #  maps an Array or Vector to a List

--- lapply( List, Function_to_apply_to_each_element )
     #  maps a List to a List 

--- sapply( List, Function_to_apply_to_each_element, simplify = "array" )
     #  maps a List to an Array
     
--- mapply( Function_to_apply_to_each_tuple_of_the_P_lists, List1, List2, ..., ListP )
     #  maps each P-tuple obtained from P Lists to a new List

###These functions are fragile and DIFFICULT TO USE -- it is not easy to get them right the first time.

```python
%%R

ddply_percentages = ddply( bnames,
                       c("year", "sex"),  #  split the data into groups, by year and sex
                       summarize,         #  reduce the values for each group
                       total_percent = sum(percent)  # compute 'total_percent' as the sum computed above
                    )   # yield a data.frame as the result
                       
cat("\nresult of tallying percentages with ddply:\n\n")
print(head(ddply_percentages))

cat("\n\result of using the 'aggregate' function in R:\n\n")
print(head(percentages))
```

#Using the right tools makes an enormous difference in EDA

##The next two commands do everything that the previous attempts above tried to do:

```python
%%R -w 800 -h 500

gender_percentages = ddply( bnames, c("year", "sex"), summarize, total_percent = sum(percent) )

print( qplot( year, total_percent, data=gender_percentages, geom="line", color=sex, main="how experts do things" ) )
```

###The plyr package has many different split-apply-combine functions:


--  <b>a_ply</b>   Split array, apply function, and discard results

--  <b>aaply</b>   Split array, apply function, and return results in an array

--  <b>adply</b>   Split array, apply function, and return results in a data frame

--  <b>alply</b>   Split array, apply function, and return results in a list 



--  <b>d_ply</b>     Split data frame, apply function, and discard results

--  <b>daply</b>     Split data frame, apply function, and return results in an array 

--  <b>ddply</b>     Split data frame, apply function, and return results in a data frame
 
--  <b>dlply</b>     Split data frame, apply function, and return results in a list



--  <b>l_ply</b>     Split list, apply function, and discard results

--  <b>laply</b>     Split list, apply function, and return results in an array

--  <b>ldply</b>     Split list, apply function, and return results in a data frame
 
--  <b>liply</b>     Experimental iterator based version of llply

--  <b>llply</b>     Split list, apply function, and return results in a list



--  <b>m_ply</b>     Call function with arguments in array or data frame, discarding results

--  <b>maply</b>     Call function with arguments in array or data frame, returning an array

--  <b>mdply</b>     Call function with arguments in array or data frame, returning a data frame
 
--  <b>mlply</b>     Call function with arguments in array or data frame, returning a list



--  <b>r_ply</b>     Replicate expression and discard results

--  <b>raply</b>     Replicate expression and return results in a array

--  <b>rdply</b>     Replicate expression and return results in a data frame
 
--  <b>rlply</b>     Replicate expression and return results in a list

###Notice that the implementation of these functions permits parallel execution, so they can scale somewhat:

```python
%%R

plyr = matrix(c(
"", "array", "data.frame", "list", "nothing",
"array", "aaply", "adply", "alply", "a_ply",
"data.frame", "daply", "ddply", "dlply", "d_ply",
"list", "laply", "ldply", "llply", "l_ply",
"n_replicates", "raply", "rdply", "rlply", "r_ply",
"function_arguments", "maply", "mdply", "mlply", "m_ply"), 6,5,byrow=TRUE)

plyr_table = data.frame( plyr[2:6,2:5] )
rownames(plyr_table) = plyr[2:6,1]
colnames(plyr_table) = plyr[1,2:5]

cat("\ninput types and output datatypes of plyr functions:\n\n")
print(plyr_table)
```

```python
%%R -w 800 -h 500

# another example:  first transform names for each year by ranking them.
# then summarize the results by taking the top name for the year.

ranked_names_by_year = ddply( bnames, .(sex, year), transform, rank = rank(-percent, ties.method="first" ))

top_name_by_year = ddply( ranked_names_by_year, .(sex, year), summarize, top = head(name,n=1L) )
                             
cat("\nfavorite boys names since 1880:\n\n"); print( unlist(subset(top_name_by_year,sex=="boy")$top) )

cat("\nfavorite girls names since 1880:\n\n"); print( unlist(subset(top_name_by_year,sex=="girl")$top) )
```

###"Quoting" column names

-- <b>.</b>                  Quote variables to create a list of unevaluated expressions for later evaluation.


```python
%%R

#  It is annoying to need to put quotes around column names all the time -- instead we can "quote" them with '.':

print( .(foo, bar) )

print( c("foo","bar") )

print( list(foo = as.symbol("foo"), bar = as.symbol("bar")) )

print( str(list(foo = as.symbol("foo"), bar = as.symbol("bar"))) )
```
