## Graphics

```python
%%R
x=rnorm(100)
y=rnorm(100)
plot(x,y)
plot(x,y,pch=20, xlab="this is the x-axis",ylab="this is the y-axis",main="Plot of X vs Y")
## pdf("Figure.pdf")
plot(x,y,pch=16, col="green")
## dev.off()
```

```python
%%R -h 800 -w 400
x=seq(1,10)
x
x=1:10
x
x=seq(-pi,pi,length=50)
y=x
print(y)

f=outer(x,y,function(x,y)cos(y)/(1+x^2))
dim(f)

opar = par(mfrow=c(1,2));
contour(x,y,f);
contour(x,y,f,nlevels=45,add=T);
par(opar)
    
```

```python
%%R
fa=(f-t(f))/2
contour(x,y,fa,nlevels=15)
image(x,y,fa)
```

```python
%%R -h 800 -w 800
opar = par(mfrow=c(3,2))
persp(x,y,fa)
persp(x,y,fa,theta=30)
persp(x,y,fa,theta=30,phi=20)
persp(x,y,fa,theta=30,phi=30)
persp(x,y,fa,theta=30,phi=60)
persp(x,y,fa,theta=30,phi=80)
par(opar)
```

## Indexing Data

```python
%%R
A=matrix(1:16,4,4)

print( A[1:2,] )
print( A[,1:2] )
print( A[1,] )
print( A[-c(1,3),] )
print( A[-c(1,3),-c(1,3,4)] )
print( dim(A) )
```
