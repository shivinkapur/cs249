#Covariance and Correlation Matrices

### Matrix normalization (obtaining centered values and z-scores)

```python
%%R
n = 3
p = 5

X <- matrix( 1:(n*p), nr=n, nc=p)
print(X)

cat("\nmatrix of column means\n")
Xbar = matrix( apply(X,2,mean), n, p, byrow=TRUE )  # fill in the matrix "by row", with repetition
print(Xbar)

cat("\nmatrix of column standard deviations\n")
Xsigma = matrix( apply(X,2,sd), n, p, byrow=TRUE )  # fill in the matrix "by row", with repetition
print(Xsigma)

cat("\nscaled version of X\n")
Zscores = (X - Xbar) / Xsigma
print(Zscores)

print(scale(X))
```

### The iris data

```python
%%R
data(iris)

print(dim(iris))
print(iris[1:5,])

X = data.matrix(iris[,1:4])
print(X[1:5,])
```

```python
%%R

cat("\ncolumn means of the iris data:\n")
print(apply(X, 2, mean))  #  column means

cat("\ncolumn standard deviations:\n")
print(apply(X, 2, sd))    #  column standard deviations

cat("\ndescriptive statistics:\n")
print( summary(iris) )
```

### Covariance Matrices

```python
%%R

n = nrow(X)
p = ncol(X)


Xbar = matrix( apply(X,2,mean), n, p, byrow=TRUE )  # fill in the matrix "by row", with repetition
Xsigma = matrix( apply(X,2,sd), n, p, byrow=TRUE )  # fill in the matrix "by row", with repetition

cat("\ncovariance matrix:\n")
print(cov(X))
print( 1/(n-1) * t( X - Xbar ) %*% ( X - Xbar ) )
```

### Correlation Matrices

```python
%%R

Zscores = (X-Xbar) / Xsigma  #  Zscores = scale(X)

cat("\ncorrelation matrix:\n")
print(cor(X))
print( 1/(n-1) * t( Zscores ) %*% ( Zscores ) )

cat("\ndiagonal of correlation matrix:\n")
print( diag(cor(X)) )
```

##Covariance and Correlation matrices are real symmetric, and positive semidefinite

```python
%%R

cat("\nThe covariance matrix is symmetric:\n")
print( cov(X) - t(cov(X)) )

cat(paste("\ndeterminant of covariance matrix =", det( cov(X) ) ))

cat("\n\nThe correlation matrix is symmetric:\n")
print( cor(X) - t(cor(X)) )

cat(paste("\ndeterminant of correlation matrix =", det( cor(X) ) ))
```

##We can decompose real symmetric matrices with their eigenvalues and eigenvectors

```python
%%R

A = matrix(c(1.4, 1.3, 1.3, 0.7), 2,2)

Aeigen = eigen(A)
Q = Aeigen$vectors
L = diag(Aeigen$values)

cat( "\nA matrix:\n" )
print( A )

cat( "\nEigendecomposition of A:\n" )

cat("\nQ:\n"); print( Q )
cat("\nL:\n"); print( L )

cat( "\nA  ==  Q * L * t(Q):\n" )
print( A )
print( Q %*% L %*% t(Q) )
```

# Interpreting the Eigendecomposition Geometrically

```python
%%R -h 400 -w 800

LinearTransformPlot = function (A = matrix(c(9/5,2/5,2/5,3/5),2,2), n = 60) {

#   The 2x2 matrix A should be real and symmetric;
#   it is used to map the unit circle (with a total of n points) to an ellipse;
#   the axes of the ellipse are the eigenvectors of A, stretched by its eigenvalues.

   t = seq(0, 2*pi, length = n)

   x = rbind( cos(t) , sin(t) )   #   x  =  a sequence of n vectors (points on the unit circle)
   Ax = A %*% x;                  #  Ax  =  the result of transforming all vectors in x by A

   x1 = x[1,];
   x2 = x[2,];

   Ax1 = Ax[1,];
   Ax2 = Ax[2,];

   A_eigen = eigen(A)
   lambda = A_eigen$values
   L      = diag(A_eigen$values)
   Q      = A_eigen$vectors

   e1 = Q[,1]  #  1st eigenvector
   e2 = Q[,2]  #  2nd eigenvector

   if (max(e1) <= 0) {
      e1 = -e1; Q[,1] = -Q[,1];  # if e is an eigenvector, then -e is an eigenvector too:
      e2 = -e2; Q[,2] = -Q[,2];  # so we choose e1 so that it has at least one positive entry.
   }

   xmax = max( lambda[1], 1.0 ) * 1.33
   ymax = max( lambda[2], 1.0 ) * 1.33

   QL = Q %*% L
   Id = diag(c(1,1))
   zeroes = c(0,0)

   row_format = " ( % 4.2f  % 4.2f ) "
   A_format = paste(" A\n ", row_format, "\n", row_format)
   A_values = sprintf( A_format, A[1,1],A[1,2], A[2,1],A[2,2] )
   QL_format = paste(" Q                       L\n ", row_format, row_format, "\n", row_format, row_format)
   QL_values = sprintf(QL_format, Q[1,1],Q[1,2],L[1,1],L[1,2], Q[2,1],Q[2,2],L[2,1],L[2,2])

   opar = par(mfrow=c(1,2), cex=0.85, cex.main=1.0, cex.sub=1.0, font.sub=2)  # make a 1x2 set of plots in one figure

   plot( x1, x2, col="red", pch=20,
      asp=1.0, xlim = c(-xmax, xmax), ylim = c(-ymax,ymax), xlab="", ylab="",
      main="transformation by A of vectors on the unit circle", sub=A_values, col.sub="blue" )
   arrows( x1, x2, Ax1, Ax2, col="green", lwd=1.0, angle=8, length=0.1 )
   arrows( zeroes, zeroes,  Q[1,],  Q[2,], col="red",  lwd=2, angle=8, length=0.1 )
   arrows( zeroes, zeroes, QL[1,], QL[2,], col="blue", lwd=2, angle=8, length=0.1 )
   points( Ax1, Ax2, col="blue", pch=20)

   plot( x1, x2, col="red", pch=20,
      asp=1.0, xlim = c(-xmax, xmax), ylim = c(-ymax,ymax), xlab="", ylab="",
      main="explanation of the transformation by A", sub=QL_values, col.sub="blue" )
   points( Ax1, Ax2, col="blue", pch=20)
   arrows( zeroes, zeroes,  Q[1,],  Q[2,], col="red",  lwd=2, angle=8, length=0.1 )
   arrows( zeroes, zeroes, QL[1,], QL[2,], col="blue", lwd=2, angle=8, length=0.1 )
   text( 1.25 *  Q[1,], 1.25 *  Q[2,], c(expression(e[1]), expression(e[2])) )
   text( 1.25 * QL[1,], 1.25 * QL[2,], c(expression(lambda[1]*e[1]), expression(lambda[2]*e[2])) )

   par(opar)
}

LinearTransformPlot()
```

```python
%%R -h 800 -w 800
#----------------------------------------------------------------------------------------
#   Exploring the effect of linear transforms on unit vectors (points on the unit circle)
#----------------------------------------------------------------------------------------

N <- 100
Thetas <- ((0:N) / N) * (2*pi)

Unit_Vectors <- rbind( sin(Thetas),
                       cos(Thetas)  )

Axes <- rbind( c( 0, 0, 1 ),
                     c( 1, 0, 0 ) )

Points <- cbind( Unit_Vectors, Axes )

plotpoints <- function(Points, Axes) {
   plot(t(Points), main=paste(deparse(substitute(Points))), xlab="x", ylab="y", pch=20,
               asp=1, col="blue", type="b" )  # asp=1 gets fixed aspect ratio = 1
   points(t(Axes), col="green", type="b", pch=20)
}


opar <- par(mfrow=c(2,2))

plotpoints(Unit_Vectors, Axes)
#----------------------------------------------------------------------------------------
#   What a rotation matrix does
#----------------------------------------------------------------------------------------

t <- pi/7
Rotation_by_Pi_over_7 <- matrix(c( cos(t), -sin(t), sin(t), cos(t) ), 2, 2)
Rotation_by_Pi_over_7

plotpoints(Rotation_by_Pi_over_7 %*% Unit_Vectors, Rotation_by_Pi_over_7 %*% Axes)

eigen(Rotation_by_Pi_over_7)

#----------------------------------------------------------------------------------------
#   What a dilation matrix does
#----------------------------------------------------------------------------------------

Dilation_by_3_1 <- diag(c( 3, 1 ))
Dilation_by_3_1

plotpoints(Dilation_by_3_1 %*% Unit_Vectors, Dilation_by_3_1 %*% Axes)

eigen(Dilation_by_3_1)

#----------------------------------------------------------------------------------------
#   What the combination of a rotation and dilation does
#----------------------------------------------------------------------------------------

A <- Rotation_by_Pi_over_7 %*% Dilation_by_3_1 %*% t(Rotation_by_Pi_over_7)
A

plotpoints(A %*% Unit_Vectors, A %*% Axes)

#----------------------------------------------------------------------------------------
#   How this relates to eigenvalues and eigenvectors
#----------------------------------------------------------------------------------------

eig <- eigen(A)
eig

eig$values[1]

(A %*% eig$vectors[,1]) / eig$vectors[,1]

lambda1 = eig$values[1]
lambda2 = eig$values[2]

e1 = eig$vectors[,1]
e2 = eig$vectors[,2]

points( matrix( c( lambda1*e1,  c(0,0), lambda2*e2 ), 3, 2, byrow=TRUE ), type="o", col="red" )

par(opar)

                                                                          
```

#The Singular Value Decomposition (SVD)

Theorem:  If $A$ is an arbitrary (real or complex) $n \times p$ matrix where $n > p$,
there is an SVD matrix decomposition
$$ A = U S V'$$
where $U$ is an $n \times n$ unitary matrix,
$V$ is a $p \times p$ unitary matrix,
and $S$ is an $n \times p$ matrix whose nonzero entries arise only on its diagonal.

```python
%%R

A = matrix(c(1.4, 1.3, 0.1, 0.7), 2,2)

Asvd = svd(A)
U = Asvd$u
S = diag(Asvd$d)
V = Asvd$v

cat( "\nA matrix:\n" )
print( A )

cat( "\nSVD of A (decomposition into U, S, V):\n" )

cat("\nU:\n"); print( U )
cat("\nS:\n"); print( S )
cat("\nV:\n"); print( V )

cat( "\nA  ==  U * S * t(V):\n" )
print( A )
print( U %*% S %*% t(V) )
```

```python
%%R -h 400 -w 1000

LinearTransformPlot = function (A = matrix(c(1.4, 1.3, 0.1, 0.7), 2,2), n = 60) {

#   The 2x2 matrix A should be real and symmetric;
#   it is used to map the unit circle (with a total of n points) to an ellipse;
#   the axes of the ellipse are the eigenvectors of A, stretched by its eigenvalues.
    
   t = seq(0, 2*pi, length=n)
   x1 = cos(t)
   x2 = sin(t)
   x = rbind( x1,  x2 )     #  x  = the unit circle  (= 2 x n matrix of (cos(t),sin(t)) vectors)

   Ax = A %*% x             #  Ax = the transformation by A of every vector in x
   Ax1 = Ax[1,]
   Ax2 = Ax[2,]

   USV = svd(A)
   sigma = USV$d
   S = diag(sigma)
   U = USV$u
   V = t(USV$v)

   u1 = U[,1]  # 1st left eigenvector
   u2 = U[,2]  # 2nd left eigenvector
   v1 = V[,1]  # 1st right eigenvector
   v2 = V[,2]  # 2nd right eigenvector
   if (max(u1) <= 0) {
      u1 = -u1; U[,1] = -U[,1];  # if u1 is an eigenvector, then -u2 is an eigenvector too:
      v1 = -v1; V[,1] = -V[,1];  # so choose u1 so that it has at least one positive entry.
   }
   if (max(u2) <= 0) {
      u2 = -u2; U[,2] = -U[,2];  # if u2 is an eigenvector, then -u2 is an eigenvector too:
      v2 = -v2; V[,2] = -V[,2];  # so choose u2 so that it has at least one positive entry.
   }

   U_sigma = U %*% diag(sigma)
   xmax = max( sigma[1], 1.0 ) * 1.33
   ymax = max( sigma[2], 1.0 ) * 1.33
   zeroes = c(0, 0)

   row_format = " ( % 4.2f  % 4.2f ) "
   A_format = paste("A\n ", row_format, "\n", row_format)
   A_values = sprintf( A_format, A[1,1],A[1,2], A[2,1],A[2,2] )
   USV_format = paste("U                       S                       V\n ",
       row_format, row_format, row_format, "\n", row_format, row_format, row_format)
   USV_values = sprintf(USV_format, U[1,1],U[1,2],S[1,1],S[1,2],V[1,1],V[1,2], U[2,1],U[2,2],S[2,1],S[2,2],V[2,1],V[2,2])

   opar = par(mfrow=c(1,2), cex=0.85, cex.main=1.0, cex.sub=1.0, font.sub=2 )   #  plot 2 figures in a 1x2 arrangement

   plot( x1, x2, pch=20, col="red",
      asp=1.0, xlim = c(-xmax, xmax), ylim = c(-ymax,ymax), xlab="", ylab="",
      main = "transformation by A of vectors on the unit circle",
      sub  = A_values, col.sub="blue" )
   arrows( x1, x2, Ax1, Ax2, col="green", lwd=1, angle=8, length=0.1 )
   points( Ax1, Ax2, pch=20, col="blue" )

   plot( x1, x2, pch=20, col="red",
      asp=1.0, xlim = c(-xmax, xmax), ylim = c(-ymax,ymax), xlab="", ylab="",
      main = "SVD explanation of the transformation by A",
      sub  = USV_values, col.sub="blue" )
   points( Ax1, Ax2, pch=20, col="blue" )
   arrows( zeroes, zeroes,       V[1,],       V[2,], col="red",   lwd=2, angle=8, length=0.1 )
   arrows( zeroes, zeroes, U_sigma[1,], U_sigma[2,], col="blue",  lwd=2, angle=8, length=0.1 )
   arrows( V[1,],  V[2,],  U_sigma[1,], U_sigma[2,], col="green", lwd=1, angle=8, length=0.1 )
   text( 1.25 *       V[1,],       1.25 * V[2,], c(expression(v[1]),          expression(v[2])         ) )
   text( 1.25 * U_sigma[1,], 1.25 * U_sigma[2,], c(expression(sigma[1]*u[1]), expression(sigma[2]*u[2])) )

   par(opar)
}

LinearTransformPlot()
```

#The Economy version of the SVD

Theorem:  If $A$ is an arbitrary (real or complex) $n \times p$ matrix where $n > p$,
there is an SVD matrix decomposition
$$ A = U S V'$$
where $U$ is an $n \times p$ prefix of a unitary matrix,
$V$ is a $p \times p$ unitary matrix,
and $S$ is an $p \times p$ matrix whose nonzero entries arise only on its diagonal.

By default, R computes this "economy" version of the SVD.  To obtain the full SVD, you must ask it to compute all eigenvectors in $U$.

```python
%%R

X = matrix( runif(5*3), 5, 3 )

Xsvd = svd(X)  # default
cat("\nU, S, and V for the default (economy) SVD\n")
print(Xsvd$u)
print(diag(Xsvd$d))
print(Xsvd$v)

Xsvd = svd(X, nu=5, nv=3 )  # we state how many eigenvectors are desired for U and V
cat("\nU, S, and V for the expensive SVD\n")
print(Xsvd$u)
print(diag(Xsvd$d))
print(Xsvd$v)
```

#Matrix Approximation with the SVD

Suppose $A$ is an $n \times p$ matrix where $n > p$, and we have the SVD matrix decomposition
$ A = U S V'.$
Assume that S = diag( \sigma_1, \ldots, \sigma_p )$ is an $n \times p$ matrix in which the only the first $k$ values on the diagonal are nonzero.

Then $A = U^{(k)} S^{(k)} V^{(k)}'$,
where
$U^{(k)}$ is an $n \times k$ matrix containing the first $k$ columns of $A$,
$S^{(k)}$ is a $k \times k$ matrix containing the upper left $k\times k$ submatrix of $S$,
and
$V^{(k)}$ is an $n \times k$ matrix containing the first $k$ columns of $V$.

```python
%%R

X = data.matrix(iris[, 1:4])
Xsvd = svd(X)

k = 2

U_2 = (Xsvd$u)[, 1:k]
S_2 = diag( (Xsvd$d)[1:k] )
V_2 = (Xsvd$v)[, 1:k]

print(dim(U_2))
print(dim(S_2))
print(dim(V_2))

Xapprox = U_2 %*% S_2 %*% t(V_2)

cat(paste( "worst-case absolute error of entries in Xapprox = ", max(abs(X - Xapprox)), "\n" ))

cat(paste( "relative error of Xapprox = ",  norm(X - Xapprox) / norm(X), "\n" ))
```
